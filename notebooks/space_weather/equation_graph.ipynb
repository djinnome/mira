```python
from sympy.vector import CoordSys3D, gradient, Del
from sympy import Symbol

# Create the symbols for the vec(Q) = - kappa * grad(T)
R = CoordSys3D('R')
nabla = Del()
kappa = Symbol('kappa')
# See
# https://docs.sympy.org/latest/modules/vector/fields.html#scalar-and-vector-fields
# for more information on the vector module
# Make temperature a scalar field
T = Symbol('T') * (R.x + R.y + R.z)
# The heat flux is a vector field
Q = - kappa * gradient(T)
print(Q)
```

```python
from IPython.display import Image
import networkx as nx
import itertools

# Put Q, kappa, T, Del as nodes in a graph and add edges between all of them
g = nx.Graph()
for u, v in itertools.combinations([Q, kappa, T, nabla], 2):
    g.add_edge(u, v)

# Draw the graph and save it to a file using graphviz
agraph = nx.nx_agraph.to_agraph(g)
agraph.draw('graph.png', prog='dot')
Image('graph.png')
```
